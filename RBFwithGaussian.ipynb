```python
import numpy as np
import pandas as pd
import random

import matplotlib.pyplot as plt
```

```python
# VIDEO referred link
# https://www.youtube.com/watch?v=Kl9xk9BukNE

# ORginal function given 
def org_func(x):
  func = np.sqrt(abs(x))*np.sin(x*(np.pi/2))
  return func
```

```python
# Centers 
X_centers = np.array([-6,-5,-2,0,1,3,5])

# dataset to be trained on
X_validate = np.linspace(-6,6,1000)

# plt.figure()
plt.plot(X_centers,org_func(X_centers),'x',markersize=15)
plt.plot(X_validate,org_func(X_validate), '--r')
```

```python
# The following line of code calculates euclidian distances of each point from every point
# Therefore, we have 7 points, we will get a 7*7 matrix of euclidian distances
# abs(X_train.reshape(-1,1) - X_train.reshape(1,-1))

def distance(x1,x2):

  dist = np.sqrt((x1.reshape(-1,1) - x2.reshape(1,-1))**2)
  return dist
```

```python
# Gaussian RBF function to convert the X vector using the centers and the width

def Gaussian_radialbase_func(Xi,Vj,sigma):
  Gaussian_func = np.exp(-(distance(Xi,Vj)**2)/(2*(sigma**2)))
  return Gaussian_func
```

```python
# Weight calculation to extract the weight matrix that would be use with Gaussian function the approximate Y
def calc_weights(X,sigma):
  Gaussian_calc = Gaussian_radialbase_func(X,X,sigma)
  Actual_output = org_func(X)
  Weights = np.linalg.solve(Gaussian_calc,Actual_output)
  return Weights
```

```python
# Approximating Y
def Interpolate_Y_RBF(X_vals,centers,sigma,Weights):

  Transform = Gaussian_radialbase_func(X_vals,centers,sigma)
  approx_output = Transform.dot(Weights)

  return approx_output
```

```python
sigma = 1
Weights = calc_weights(X_centers,sigma)

# For Training data
Y_approx = Interpolate_Y_RBF(X_validate,X_centers,sigma,Weights)
Y_actual = org_func(X_validate)

# For approximating 
X = np.array([3.6])
Y_real = org_func(X)
Y_calc = Interpolate_Y_RBF(X,X_centers,sigma,Weights)

print("the real val = ", Y_real)
print("the Y val estimated from RBF = ",Y_calc)


# print(Y_approx)
# print(Y_actual)
# plt.figure(figsize=(12,6))
plt.plot(X_centers,org_func(X_centers),'x',markersize=15)
plt.plot(X_validate,Y_actual, '--r')
plt.plot(X_validate,Y_approx, '--b')
plt.plot(X,Y_calc, 'x',markersize=25,color="blue")
plt.plot(X,Y_real, 'x',markersize=25,color="red")
```
